```python
from ultralytics import YOLO

# Load model pre-trained weights
model = YOLO('/kaggle/input/yolo/pytorch/default/1/best.pt')
```

```python
yaml_content = """
path: /kaggle/input/home-fire-dataset
train: train/images
val: val/images
test: test/images

nc: 2
names: ['fire', 'smoke']
"""

with open("/kaggle/working/home-fire.yaml", "w") as f:
    f.write(yaml_content)
```

```python
model.train(
    data="home-fire.yaml",
    epochs=200,          # maximum possible epochs
    patience=20,         # stop if no improvement for 20 epochs
    imgsz=640,
    batch=8,
    device=0
)
```

```python
model = YOLO("/kaggle/input/last/pytorch/default/1/best (4).pt")

# Evaluate on the test split
metrics = model.val(
    data="/kaggle/working/home-fire.yaml",
    split="test"
)

print(metrics)
```
